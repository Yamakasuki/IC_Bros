# Importando as Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python

# Lista com os nomes dos arquivos CSV
arquivos_csv = [
    "../../data/Volt(8)/10012025-114153.csv",
    "../../data/Volt(8)/10012025-114254.csv",
    "../../data/Volt(8)/10012025-114353.csv",
    "../../data/Volt(8)/10012025-114448.csv",
    "../../data/Volt(8)/10012025-114549.csv"
]

for arquivo in arquivos_csv:
    # Carregar o CSV
    df = pd.read_csv(arquivo)

    # Criar a coluna de tempo reiniciando para cada arquivo
    df["Tempo (s)"] = np.arange(len(df)) / 30

    # Definir a coluna de tempo como índice
    df.set_index("Tempo (s)", inplace=True)

    # Criar gráfico dos ângulos do quadril e joelho
    plt.figure(figsize=(20, 6))
    sns.lineplot(x=df.index, y=df.iloc[:, 19], color='red', label="Quadril X")
    sns.lineplot(x=df.index, y=df.iloc[:, 20], color='green', label="Quadril Y")
    sns.lineplot(x=df.index, y=df.iloc[:, 21], color='blue', label="Quadril Z")
    sns.lineplot(x=df.index, y=df.iloc[:, 22], color='red', linestyle="--", label="Joelho X")
    sns.lineplot(x=df.index, y=df.iloc[:, 23], color='green', linestyle="--", label="Joelho Y")
    sns.lineplot(x=df.index, y=df.iloc[:, 24], color='blue', linestyle="--", label="Joelho Z")

    # Configuração do gráfico
    plt.title(f"Ângulos do Quadril e Joelho no Tempo - {arquivo.split('/')[-1]}", fontsize=16, fontweight="bold")
    plt.xlabel("Tempo (s)", fontsize=14)
    plt.ylabel("Amplitude do Sinal", fontsize=14)
    plt.legend()
    plt.show()
```

# Usando o tempo total para tentativa de uma analise

```python
# Lista com os nomes dos arquivos CSV
arquivos_csv = [
    "../../data/Volt(8)/10012025-114153.csv",
    "../../data/Volt(8)/10012025-114254.csv",
    "../../data/Volt(8)/10012025-114353.csv",
    "../../data/Volt(8)/10012025-114448.csv",
    "../../data/Volt(8)/10012025-114549.csv"
]

tempo_total = 0  # Para ajustar a linha do tempo corretamente

for arquivo in arquivos_csv:
    df = pd.read_csv(arquivo)

    # Criar a coluna de tempo acumulando os tempos entre os arquivos
    df["Tempo (s)"] = np.arange(len(df)) / 30 + tempo_total
    tempo_total = df["Tempo (s)"].iloc[-1]  
    df.set_index("Tempo (s)", inplace=True)

    # Criar gráfico dos ângulos do quadril
    plt.figure(figsize=(20, 6))
    sns.lineplot(x=df.index, y=df.iloc[:, 18], color='red', label="Quadril X")
    sns.lineplot(x=df.index, y=df.iloc[:, 19], color='green', label="Quadril Y")
    sns.lineplot(x=df.index, y=df.iloc[:, 20], color='blue', label="Quadril Z")
    # Criar gráfico dos ângulos do joelho
    sns.lineplot(x=df.index, y=df.iloc[:, 21], color='red', linestyle="--", label="Joelho X")
    sns.lineplot(x=df.index, y=df.iloc[:, 22], color='green', linestyle="--", label="Joelho Y")
    sns.lineplot(x=df.index, y=df.iloc[:, 23], color='blue', linestyle="--", label="Joelho Z")

    # Configuração do gráfico
    plt.title(f"Ângulos do Quadril e Joelho no Tempo - {arquivo.split('/')[-1]}", fontsize=16, fontweight="bold")
    plt.xlabel("Tempo (s)", fontsize=14)
    plt.ylabel("Amplitude do Sinal", fontsize=14)
    plt.legend()
    plt.show()
```

# Conclusões

Ao longo deste teste, foi possível realizar uma análise detalhada dos ângulos do joelho e do quadril, tanto de forma individual quanto sincronizada. A separação clara dos eixos X, Y e Z permitiu uma avaliação mais precisa do comportamento de cada articulação ao longo do tempo. Além disso, identificou-se que os eixos X apresentaram maior oscilação, dificultando sua utilização na análise de compensação de movimento, enquanto os eixos Y e Z se mostraram mais estáveis e úteis para identificar variações. Dessa forma, a abordagem utilizada possibilitou uma melhor compreensão dos padrões de movimento e abre caminho para futuras investigações e aprimoramentos na análise biomecânica.
