# Importando as Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

# Importando Dataset

```python
# Apagando a primeira linha para tratamento de hz
# Pegando apenas a 23 coluna para fazer o data frame 
# Adicionando um Header para a coluna  

df = pd.read_csv("../../data/Data_set.csv",skiprows=1 ,usecols=[22],header=None) 
df.columns =["STS_JOELHO_X"]  
print(df.head())
```

```python
df.info()
print(df.isnull().sum())  # Quantidade de valores nulos
df.describe()
# Descrição estatistica da coluna
```

```python
# Contar valores ausentes
print("Valores nulos:", df.isnull().sum())

# Contar valores duplicados
print("Valores duplicados:", df.duplicated().sum())
```

# Tentativa de Criação de Index temporal

```python
# Criando uma coluna de tempo (considerando que a 1ª linha é o tempo 0s)]

# Definindo essa coluna como o índice do DataFrame

df["Tempo (s)"] = np.arange(len(df)) / 30

df.set_index("Tempo (s)", inplace=True)

# Visualizando os dados após a conversão
print(df.head())

#Usando isso podemos ver que o exercicio vai ter no Total de .... 
print(df.tail())
```

# Importando Bibliotecas para Visualização de Dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20, 6))
plt.plot(df.index, df["STS_JOELHO_X"], color='blue', alpha=0.7, linestyle="-")

# Configurações do gráfico
plt.title("STS_JOELHO_X no Tempo", fontsize=16, fontweight="bold")
plt.xlabel("Tempo (s)", fontsize=14)
plt.ylabel("Amplitude do Sinal", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["STS_JOELHO_X"], bins=100, kde=True, color="blue")

# Configurações
plt.title("Distribuição dos Valores do Sinal", fontsize=14)
plt.xlabel("Amplitude do Sinal", fontsize=12)
plt.ylabel("Frequência", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df["STS_JOELHO_X"], color="blue")

plt.title("Boxplot do Sinal Sensorial", fontsize=14)
plt.ylabel("Amplitude do Sinal", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
print(df.columns)  # Mostra todas as colunas disponíveis
```

```python
df["Média_Móvel"] = df["STS_JOELHO_X"].rolling(window=30).mean()

plt.figure(figsize=(20, 6))
plt.plot(df.index, df["STS_JOELHO_X"], color='red', alpha=0.1, label="Sinal Original")
plt.plot(df.index, df["Média_Móvel"], color='blue', linestyle="-", label="Média Móvel (1s)")

plt.title("Sinal Sensorial com Média Móvel (1s)", fontsize=16, fontweight="bold")
plt.xlabel("Tempo (s)", fontsize=14)
plt.ylabel("Amplitude do Sinal", fontsize=14)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df)
plt.title("Evolução Temporal do STS ")
plt.xlabel("Tempo")
plt.ylabel("Angulação")
plt.show()
```
